```python
import requests
import time

def send_data(food_weight, water_weight):
    server_name = "nglam.xyz"
    server_port = "80"
    server_path = "/api/v2/sensor"

    if requests.get('http://www.google.com'):
        print("Connection successful!")
        url = "http://" + server_name + ":" + server_port + server_path
        params = {'food_weight': food_weight, 'water_weight': water_weight}
        response = requests.post(url, params=params)
        
        if response.status_code == 200:
            print(response.text)
        else:
            print("Request failed.")
    else:
        print("WiFi Disconnected")
```

```python
#Convert Unix timestamp to human readable format
def convert_time(unix_time):
    return time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(unix_time))

#Get the current time in Unix timestamp
def get_time():
    return int(time.time())
```
